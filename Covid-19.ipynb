# Covid-19 Data Transformation for further analysis
## Data forked from Novel Coronavirus (COVID-19) Cases, provided by JHU CSSE 
## (https://systems.jhu.edu/research/public-health/ncov/) 
### GitHub Repository: https://github.com/CSSEGISandData/COVID-19

```python
# Import Statements
import sys
import csv
import time
import numpy as np
import pandas as pd
from pandas import DataFrame
from datetime import datetime
```

```python
# Import Confirmed Cases file
dfc = pd.read_csv('~\\Documents\\PythonUsersMeetUps\\COVID-19-master\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_time_series\\time_series_19-covid-Confirmed.csv')
df1c = dfc.iloc[:,1:].drop(['Lat','Long'], 1).rename(columns={"Country/Region": "Country"}).groupby('Country').sum().T
df1c
```

```python
# Import Death Cases file
dfd = pd.read_csv('C:\\Users\\InfiniteIntelligence\\Documents\\PythonUsersMeetUps\\COVID-19-master\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_time_series\\time_series_19-covid-Deaths.csv')
df1d = dfd.iloc[:,1:].drop(['Lat','Long'], 1).rename(columns={"Country/Region": "Country"}).groupby('Country').sum().T
df1d
```

```python
# Import Recovered Cases file
dfr = pd.read_csv('C:\\Users\\InfiniteIntelligence\\Documents\\PythonUsersMeetUps\\COVID-19-master\\COVID-19-master\\csse_covid_19_data\\csse_covid_19_time_series\\time_series_19-covid-Recovered.csv')
df1r = dfr.iloc[:,1:].drop(['Lat','Long'], 1, inplace=False).rename(columns={"Country/Region": "Country"}).groupby('Country').sum().T
df1r
```

```python
# Select a few Relevant Countries Only 
cols_to_keep = ['Australia', 'Austria', 'Belgium', 'China', 'Denmark', 'Germany', 'Greece', 'Ireland', 'Israel', 'Italy',
                'Japan', 'Korea, South', 'Netherlands', 'Portugal', 'Slovenia',
                'Spain',  'Sweden', 'Switzerland', 'US', 'United Kingdom']

df1c = df1c.drop(df1c.columns.difference(cols_to_keep), axis=1)
df1d = df1d.drop(df1d.columns.difference(cols_to_keep), axis=1)
df1r = df1r.drop(df1r.columns.difference(cols_to_keep), axis=1)

# Check Selection worked
set(df1c.columns)
```

```python
# Melt Dataset to shape it differently
df1c = df1c.reset_index().melt(id_vars='index', var_name='Country', value_name='Confirmed', col_level=None)
df1d = df1d.reset_index().melt(id_vars='index', var_name='Country', value_name='Deaths', col_level=None)
df1r = df1r.reset_index().melt(id_vars='index', var_name='Country', value_name='Recovered', col_level=None)
```

```python
# Merge 3 Datasets 
df = pd.merge(df1c,df1d,on=['index','Country']).merge(df1r,on=['index','Country'])
df
```

```python
# Set Datetime Format on 'index' column
df['index'] = pd.to_datetime(df['index'], format= '%m/%d/%y', infer_datetime_format=True)

# OPTIONAL: df = df.set_index('index')
```

```python
# Create New KPIs 
df['Closed_Cases'] = df['Deaths']+df['Recovered']
df['Active_Cases'] = df['Confirmed']-df['Closed_Cases']
df['Active_for_1_Closed'] = df['Active_Cases']/df['Closed_Cases']
df
```

```python
# Replace 'NaN' and 'inf' values 
df['Active_for_1_Closed'].replace([np.inf, np.nan], 0, inplace=True)
```
